```python
from matplotlib import pyplot as plt
import random
x =[i for i in range(1,101)]
y =[]
for i in range(100):
    y.append(random.randint(1,101))
#colors: w r y g k c b m 
#shapes: o p s d h x v . , 1 2 3 4 8 - + | : < > -- -.
plt.plot(x,y, 'rh')
plt.title('Some Title')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.show()
```

```python
from matplotlib import pyplot as plt
import random
x =[i for i in range(1,101)]
y =[]
for i in range(100):
    y.append(random.randint(1,101))
#colors: w r y g k c b m 
#shapes: o p s d h x v . , 1 2 3 4 8 - + | : < > -- -.
lines = plt.plot(x,y)
plt.setp(lines, linewidth=3, color='g')
plt.show()
```

```python
from matplotlib import pyplot as plt
histogram_dataset =[1,4,6,8,5,2,1,4,6,7,4,2,3,7,9,0]
plt.hist(histogram_dataset)
plt.title('Histogram')
plt.xlabel('Numbers (1-10)')
plt.ylabel('count')
plt.hist(histogram_dataset, color='m', cumulative=False, density=True)
plt.show()
```

```python
from matplotlib import pyplot as plt

bar_data = {'key': 6, 'gold': 2, 'silver': 9, 'gear': 23}
mydata =plt.bar(bar_data.keys(), bar_data.values())
plt.setp(mydata, color='g')
plt.show()

hist_data = [1,4,6,8,9,0]
plt.hist(hist_data)
plt.title('Histogram title')
plt.xlabel('x-value')
plt.ylabel('y-value')
plt.hist(hist_data, color='r', cumulative = 'false', density='true')
plt.show()
```

```python
#piechart
from matplotlib import pyplot as plt
piechart_data={'key': 6, 'gold': 2, 'silver': 9, 'gear': 23}
explode =(0.2,0,0,0) #explode helps when you want to put more empasis on a specific type of data either in presentation
plt.pie(piechart_data.values(), labels=piechart_data.keys(), startangle =180, explode=explode, shadow=True)
plt.show()
```
