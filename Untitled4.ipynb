```python
import numpy as np
from scipy import linalg
import scipy.misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#creating input array
a =np.array([[1,2,-3], [2,-5,4], [4,4,-1]])
#solution array
b =np.array([[-3], [13], [5]])
x =linalg.solve(a, b)
print(x)
```

Scikit-Learn is a library in python that provides many unsupervised and supervised learning algorithms.


```python
#import inbuilt dataset and libraries
from sklearn import datasets
from sklearn import metrics #used to determine how the model is doing
from sklearn.tree import DecisionTreeClassifier #import an algorithm

#load iris datasets
dataset =datasets.load_iris()

#create a model
model = DecisionTreeClassifier()
model.fit(dataset.data, dataset.target) #its going to learn from dataset

```

```python
#make predictions
expected =dataset.target
predicted =model.predict(dataset.data)
predicted #show predictions
```

```python
#This is the original target
dataset.target
# expected
```

```python
#lets see how the model is doing
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```
