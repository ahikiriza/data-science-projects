# Spliting the data

```python
# we use 70% for training and the rest 30% for testing
from sklearn.model_selection import train_test_split
x =df.drop('class', axis=1)
y =df['class']
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.3, random_state=1)#setting the randomstate to generate random variables once
```

```python
x_train
```

```python
# y_train
#check the split of data
print("{}% data is in the training set".format((len(x_train)/len(df))*100))
print("{}% data is in the test data".format((len(x_test)/len(df))*100))
```

# Data preparation

```python
#check hidden misssing values
x_train.head()
# see alot of zeros
```

# Lets replace zeros with mean

```python
from sklearn.impute import SimpleImputer

replace_ =SimpleImputer(missing_values =0, strategy="mean")

cols =x_train.columns

x_train=pd.DataFrame(replace_.fit_transform(x_train))
x_test=pd.DataFrame(replace_.fit_transform(x_test))

x_train.columns =cols
x_test.columns =cols
```

```python
x_train.head()
```

# Logistic Regression Model


```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegression

#fit the model on train
model =LogisticRegression(solver="liblinear")
model.fit(x_train, y_train)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
#predict on test
y_predict =model.predict(x_test)
y_predict
```

```python
#check the score
model_score= model.score(x_test, y_test)
print(model_score)
```

```python
#build confusion matrix
cm =metrics.confusion_matrix(y_test, y_predict, labels=[1,0])

df_cm =pd.DataFrame(cm, index = [i for i in ["1", "0"]],
                        columns =[i for i in ["predict 1", "predict 0"]])
plt.figure(figsize=(7,5))
sns.heatmap(df_cm, annot=True)
```

from above, predicted 48 correctly and 37 wrongly for 1(diabetic)
also predicted 14 wrongly and 1300 correctly for 0(non-diabetic)
To use your model in production, u need to be in 90s

# classification Report

```python

from sklearn.metrics import classification_report

print("classification Report")
print(metrics.classification_report(y_test, y_predict, labels=[1, 0]))

#f1-score gives the weighted average of the precision and recall
```
